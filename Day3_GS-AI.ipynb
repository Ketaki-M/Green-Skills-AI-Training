```python
import matplotlib.pyplot as plt
```

```python
months=['Jan','Feb','Mar','Apr','May','Jun']
energy_consumption=[1200,1300,1100,1500,1400,1600]
```

```python
plt.plot(months,energy_consumption,marker='o',color='r',linestyle='--')
plt.title('Energy Consumption over 6 months')
plt.xlabel('Months')
plt.ylabel('Energy Consumption')
plt.show()
```

```python
plt.bar(months,energy_consumption,color='r')
plt.title('Energy Consumption over 6 months')
plt.xlabel('Months')
plt.ylabel('Energy Consumption')
plt.show()
```

```python
plt.pie(energy_consumption,autopct='%1.1f%%',labels=months)
plt.title('ENERGY CONSUMPTION OVER MONTHS')
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.boxplot(df)
plt.show()
```
